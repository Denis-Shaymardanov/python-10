# f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30
1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0

```python
#график функции
from sympy import *
from sympy.plotting import plot
from matplotlib import *
init_printing()
x = Symbol('x')
f = -12*x*x*x*x* sin(cos(x)) - 18*x*x*x + 5*x*x + 10*x - 30
plot(f,(x,-50,50), line_color = 'red',title = 'График функции f')
```

```python
#Корни уравнения
solveset(Eq(f,0),x)
```

```python
#Возрастание и убывание функции
df = f.diff(x)
print('Ha участках, где: \n(df < 0) - функция убывает, \n(df > 0) - функция возрастает')
plot(df,(x, -20, 20), line_color ='blue', title = 'График функции df')
plot(f,(x, -20, 20),line_color ='green', title = 'График функции f')
```

```python
#Производная
print(df)
```

```python
#Точка пересечения и минимальное значение функции
print(f'Точка пересечения с осью Y: [0, {f.subs(x,0)}]')
print(f'Функция f(x) = {f} имеет множество корней и не имеет экстремумов')
```
